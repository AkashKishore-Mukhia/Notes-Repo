## Decorator

Decorator is a function which can take a function as a argument and extends it's functionality and returns modified function with expected functionality.

input function --------------->Decorator----------output functoin with extended functionality----------> 

The main objective of decorator functions is we can extend the functionality of existing functions without modifies that function.
The decor function returns a inner fucntion which is same concept as closure.

```python
def greet(name):
    print('hello', name)

greet('john')
greet('smith')
```

But we want to modify this function to provide different message if name is Sunny.
We can do this without touching wish() function by using decorator.

```python
def decor(func):
    def inner(name):
        if name == 'sam':
            print('hello mr:', name)
        else:
            func(name)
    return inner
```

```python
decorfun = decor(greet)
decorfun('john')
decorfun('smith')
decorfun('sam')
```

```python
# above way of calling decorfun is identical as below
@decor
def greet(name):
    print('hello', name)
```

```python
greet('john')
greet('smith')
greet('sam')
```

```python
def smart_div(func):
    def inner(a, b):
        if b == 0:
            print('Opps can not divide')
        else:
            print('reuslt', a/b)
    return inner
```

```python
def division(a, b):
    print('result', a/b)
```

```python
# Without decorator we will get error
division(20, 2)
division(20, 0)
```

```python
@smart_div
def division(a, b):
    print('result', a/b)
division(20, 2)
division(20, 0)
```

### Decorator Chaning
We can define multiple decorators for the same function and all these decorators will form Decorator Chaining.

```python
def decor1(func):
    def inner(a):
        x = func(a)
        return x*x
    return inner
```

```python
def decor(func):
    def inner(a):
        x = func(a)
        return 2*x
    return inner
```

```python
@decor1
@decor
def f(a):
    return a

print(f(2))
print(f(4))
```
