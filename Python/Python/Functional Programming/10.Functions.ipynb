# Fucntions

- If a group of statement is repeatedly required then it is not recommended to write these statements everytimes separetly. We have to define these statements as a single unit and can call any number of times based on our requirements without rewriting. This unit is nothing but function.
- The main adv of function is code reusability.
- Python supports two type of function
    - built-in function
    - user defined function


### Syntax to create user defined function :
```python
def function_name(parameters):
    """doc-string"""
    ------
    ------
    return Value
```

## Parameters

Parameters are input to the function. If a fucntion contains parameters, then at the time of calling, compulsory we should provide values, otherwise we will get error.

## Return statement

Function can take input values as parameters and execute businees logic, and returns output to the caller with return statement.

`Note:` In python a function can return any number of values.

## Type of Arguments

There are 4 types of arguments are allowed in python:
- Positional Arguments
- Keyword Arguments
- Default Arguments
- Variable Length Arguments

### 1. Positional Arguments:
- These are the arguments passed to function in correct positional order.

````python
def sub(a, b):
    print(a-b)
````

````python
sub(200, 100)
sub(100, 200)
````

### 2. Keyword Arguments
- We can pass arguments values by keywords i.e by parameter name

````python
sub(a=200, b=100)
sub(b=100, a=200)
````

`Note:` We can take both positional and keyword arguments simultaneouly. But first we have to take positional argument and then keyword argument, otherwise we will get syntaxerror.

### 3. Default Arguments
- Sometimes we can provide default values for our positional arguments.

````python
def wish(name='john'):
    print("happy birthday", name)
````

````python
wish()
wish('mark')
````

`Note:` After default arguments we should not take non default arguments.

### 4. Variable Length Arguments
- Sometimes we can pass variable number of arguments to our function, such type of arguments are called variable length arguments.
- We can declare a variable length arguments with * symbol.
- Internally all these values represented in the form of tuple.

````python
def variable_length_arg(*vLenthArg):
    print(vLenthArg)
````

````python
variable_length_arg('john', '23', 'fc3453')
````

`Note:` We can mix variable lenth variable with positional arguments

````python
def vla_with_pos(name, *vLenthArg):
    print(name)
    print(vLenthArg)

vla_with_pos('john', '23', 'fc3453')
````

`Note:` After variable length argument, if we are taking any other arguments then we should provide values as keyword arguements.

````python
def vla_with_key(*vLenthArg, id):
    print(vLenthArg)
    print(id)

vla_with_key('john', '23', id='fc3453')
````

`Note:` We can declare key word variable length arguments also.

````python
def vlka(**vLengthKeyArg):
    print(vLengthKeyArg)
````

````python
vlka(name='john', age=23, id='fc3453')
````

## Types of Variable
Python supports 2 types of variables:
1. Global Variable
2. Local Variable

### Global Keyword:
We can use global keyword for the following 2 purposes:
1. To declare global varibale inside function
2. To make global variable available to the function, so that we can perform required modifications

````python
a = 10
def f1():
    # it will not effect the global variable 'a'
    a = 777
    print(a)

def f2():
    print(a)
````

````python
f1()
f2()
````

````python
def f1():
    # declare global varibale inside function
    global a
    a = 10
    print(a)

def f2():
    print(a)
````

````python
f1()
f2()
````

````python
a = 10
def f1():
    # make global variable available to the function
    # global a
    a = 777
    print(a)
def f2():
    print(a)
````

````python
f1()
f2()
````

`Note:` If global variable and local variable having the same name then we can access global variable inside a function as follows.

````python
a = 10
def f1():
    a = 777
    print(a)
    print(globals()['a'])
````

````python
f1()
````

## Lambda Function
Sometimes we can declare a function without any name,such type of nameless functions are called anonymous functions or lambda functions.

### Syntax to create Lambda function

```python
    # lambda argument_list : expression
    lambda n: n * n
```

````python
s = lambda a,b: a+b
b = lambda a,b:a if a > b else b
````

````python
print(s(10, 20))
b(10, 40)
````

## filter() Function
We can use `filter` function to filter values from the given sequence based on some condition.
```python
    filter(function, sequence)
    # where fucntion argument is responsible to perform conditional check on sequence, sequence can be List or Tuple or String.
```


````python
l = [1,2,3,4,5,6,7,8,9]
list(filter(lambda x: x%2 == 0, l))
````

## Map() Function
For every element present in the given sequence, apply some functionality and generate new element with the required modification. For this requirement we should go for `map` function.

```python
    map(function, sequence)
    # where fucntion argument is responsible for generate element with required modification on sequence, sequence can be List or Tuple or String.
```

````python
def double(x):
    return x*x
````

````python
l = [1,2,3,4,5,6,7,8,9]
list(map(lambda x:x*x, l))
````

````python
list(map(double, l))
````
